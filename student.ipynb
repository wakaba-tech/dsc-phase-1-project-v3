```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the datase of Aviation data
df = pd.read_csv('data/Aviation_Data.csv', low_memory=False)
df
```

```python
#looking for how many columns and rows we have in our dataset
df.shape
```

```python
# checking all the columns that we have in our data
df.columns
```

```python
# looking if we have duplicated data in our data set
df.duplicated()
```

```python
#checking if we have duplicated values in our dataset
df.drop_duplicates(inplace=True)
```

```python
# getting the information of the data and checking if we have missing values
df.info()
```

```python
#in our dataset we are looking for risk invovled and here i dropped the columns that are not relevant with our findings
columns_to_drop = [
    'Airport.Code',
    'Airport.Name',
    'Schedule',
    'Air.carrier',
    'Latitude',
    'Longitude',  
    'FAR.Description']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# checking if the columns we have droped are still in our data.
df.shape
```

```python
# checking the percentange of missing values in every column to determine if we need to retain or drop them for our project
df.isnull().mean()*100
```

```python
#code used to find mean, mode, median, standard deviation, first and third quantile of our data set
df.describe()
```

```python
#dropping the rows with less tha 10% missing values 
df.dropna(subset=['Location', 'Country', 'Engine.Type', 'Purpose.of.flight', 'Weather.Condition', 'Report.Status', 'Injury.Severity', 'Aircraft.damage', 'Registration.Number', 'Make', 'Model', 'Amateur.Built'], inplace=True)
```

```python
# filing the missing data and i used mean because the engines are numerical data and also they use integers numbers 
df['Number.of.Engines'].fillna(df['Number.of.Engines'].mean(), inplace=True)
df['Total.Fatal.Injuries'].fillna(0,inplace=True)
df['Total.Serious.Injuries'].fillna(0,inplace=True)
df['Total.Minor.Injuries'].fillna(0,inplace=True)
df['Total.Uninjured'].fillna(df['Total.Uninjured'].median(), inplace=True)
```

```python
#as much as it missed alot of data,i felt its of importance as it will let me understand which aircraft category caused alot of accidents
# since this is a categorical data i have decided to use unknown to replace the missing data because they cannot be affecte by mean, median, mode
df['Aircraft.Category'].fillna('unknown', inplace=True)
df['Broad.phase.of.flight'].fillna('unknown', inplace=True)
df['Publication.Date'].fillna('unknown', inplace=True)
```

```python
# checking the number of misiing values in each column
df.isna().sum()
```

```python
#code to check if we still have missing values in the dataset
df.info()
```

```python
#saving the cleaned dataset
df.to_csv('cleaned_Aviation_data.csv', index=False)
df.to_csv('cleaned_Aviation_data.csv', index=False)
```

```python
# graph to show the areas that accident have happened more 
location_accidents = df['Location'].value_counts().reset_index()
location_accidents.columns = ['Location', 'Accident Count']
location_accidents.sort_values(by='Accident Count', ascending=False, inplace=True)
plt.figure(figsize=(20,7))
plt.barh(location_accidents['Location'].head(10), location_accidents['Accident Count'].head(10), color='lightblue')
plt.xlabel('Number of Accidents')
plt.title('Top 10 Locations with Most Accidents')
plt.gca().invert_yaxis()
plt.show()
```

```python
sector = df['Purpose.of.flight'].value_counts()
```

```python
# a bargraph to show the relationship between purpose of flight and accident
fig,ax = plt.subplots(figsize = (15,10))
fig.set_tight_layout(True)
ax.bar(sector.index.tolist()[:10],sector.tolist()[:10])
ax.set_xlabel('Purpose Of Flight')
ax.set_ylabel('Number of accidents and incidents')
ax.set_title('Top 10 most affected sectors in the industry')
plt.show();
```

```python
#looking for which engine type that does not requre sub egines for it to run efficiently
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Engine.Type', y='Number.of.Engines')
plt.title('Scatter Plot of Engine Type vs. Number of Engines')
plt.xlabel('Engine Type')
plt.ylabel('Number of Engines')
plt.xticks(rotation=90)
plt.show()
```

```python
#looking for which make that has resulted to many accident 
accidents = df['Make'].value_counts().reset_index()
accidents.columns = ['Make', 'Accident Count']
accidents.sort_values(by='Accident Count', ascending=False, inplace=True)
plt.figure(figsize=(20,7))
plt.barh(accidents['Make'].head(10), accidents['Accident Count'].head(10), color='blue')
plt.xlabel('Make')
plt.title('Top 10 Locations with Most Accidents')
plt.gca().invert_yaxis()
plt.show()
```

```python
# looking for the engine type that has caused mor accident and caused more fatal injuries
top_30_engines = df['Engine.Type'].value_counts().nlargest(30).index
filtered_df =df[df['Engine.Type'].isin(top_30_engines)]
plt.figure(figsize=(12, 8))
sns.scatterplot(data=filtered_df, x='Engine.Type', y='Total.Fatal.Injuries')

plt.title('Scatter Plot of Top 30 Engine Types vs. Total Fatal Injuries')
plt.xlabel('Engine.Type')
plt.ylabel('Total.Fatal.Injuries')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#looking for the rate at which accident happen in the last 20 years
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Event.Date'].dt.year
df['Month'] = df['Event.Date'].dt.month
recent_years = df['Year'].max() - 20
df_recent = df[df['Year'] > recent_years]

accidents_per_month = df.groupby(['Year', 'Month']).size().unstack(fill_value=0)
plt.figure(figsize=(16, 12))
sns.heatmap(accidents_per_month, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Accident Frequency over Time(Last 20 Years)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```
